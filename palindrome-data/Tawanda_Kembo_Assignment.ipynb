```python
# 0: Add the usual suspects
import pandas as pd
```

```python
# 1: Download the dataset
file_url = "https://ndownloader.figshare.com/files/14443145"
raw_data = pd.read_excel(io=file_url,header=1,index_col=0)
df = pd.DataFrame(raw_data)
df.head()
```

```python
# 2(a): Total number of people living with HIV
print(df["NoPLHIV"].sum())
```

```python
# 2(b): average NoPLHIV of the two estimates used for “Xhariep”
two_b_df = df[df.Code == "DC16"]
print(two_b_df["NoPLHIV"].mean())
```

```python
# 2(c): Number of people not living iwth HIV (Assuming here that 'Prevalence_% percentage of people in NoPLHIV')
df["Total_Two_C"] = ((100 - df["Prevalence_%"])/ df["Prevalence_%"]) * df["NoPLHIV"]
df.head()
```

```python
# 2 (d) total NoPLHIV in all the cities (districts with the word “city” or “metro” in the name)
frames = [df.filter(like="Metro", axis=0),df.filter(like="City", axis=0)]
two_d_df = pd.concat(frames)
print(two_d_df["NoPLHIV"].sum())
```
